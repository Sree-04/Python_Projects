# CALCULATOR

```python
import tkinter
from tkinter import *
```

```python
expression = " "
def click(num):
    global expression
    expression +=num
    text1.config(text=expression)
    
def equal_click():
    global expression
    total=" "
    
    if expression!=" ":
        
        try:
            total = eval(expression)

        except:
            total= "error"
            expression = " "
            
    text1.config(text=total)
    
def clear():
    global expression
    expression = " "
    text1.config(text= " ")
```

```python
window = Tk()

window.title("CALCULATOR")

window.geometry("400x200")

text1= Label(window,text=" " ,width=40)
text1.grid(columnspan=4,ipadx=60)

button1=Button(window, text="1",command=lambda:click("1"), bg="black", fg="red")
button1.grid(row=1,column=0)

button2=Button(window, text="2",command=lambda:click("2"), bg="black", fg="red")
button2.grid(row=1,column=1)

button3=Button(window, text="3",command=lambda:click("3"), bg="black", fg="red")
button3.grid(row=1,column=2)

button4=Button(window, text="4",command=lambda:click("4"), bg="black", fg="red")
button4.grid(row=2,column=0)

button5=Button(window, text="5",command=lambda:click("5"), bg="black", fg="red")
button5.grid(row=2,column=1)


button6=Button(window, text="6",command=lambda:click("6"), bg="black", fg="red")
button6.grid(row=2,column=2)


button7=Button(window, text="7",command=lambda:click("7"), bg="black", fg="red")
button7.grid(row=3,column=0)

button8=Button(window, text="8",command=lambda:click("8"), bg="black", fg="red")
button8.grid(row=3,column=1)


button9=Button(window, text="9",command=lambda:click("9"), bg="black", fg="red")
button9.grid(row=3,column=2)


button10=Button(window, text="0",command=lambda:click("0"), bg="black", fg="red")
button10.grid(row=4,column=1)

button11=Button(window, text="clear",command=lambda:clear(), bg="black", fg="red")
button11.grid(row=4,column=0)

button12=Button(window, text="=",command=lambda:equal_click(), bg="black", fg="red")
button12.grid(row=4,column=2)

button13=Button(window, text="+",command=lambda:click("+"), bg="black", fg="red")
button13.grid(row=1,column=3)


button14=Button(window, text="-",command=lambda:click("-"), bg="black", fg="red")
button14.grid(row=2,column=3)


button15=Button(window, text="/",command=lambda:click("/"), bg="black", fg="red")
button15.grid(row=3,column=3)

button16=Button(window, text="*",command=lambda:click("*"), bg="black", fg="red")
button16.grid(row=4,column=3)

button17=Button(window, text=".",command=lambda:click("."), bg="black", fg="red")
button17.grid(row=5,column=1)


window.mainloop()
```
